```python
import pandas as pd
pd.options.display.max_rows = 4000

FILE_NAMES = ['../Experiments/exp5_d4_data.csv', '../Experiments/exp5_d5_data.csv']

df_list = []

for file_name in FILE_NAMES:
    df = pd.read_csv(file_name)
    df = df = df[df['depth'].notna()] # Remove empty rows 
    df_list.append(df)
    
merged_df = pd.concat(df_list, ignore_index=True)
merged_df.columns = [c.replace(' ', '_') for c in merged_df.columns]
merged_df.drop(merged_df[merged_df.move_number == 0].index, inplace=True)
merged_df
```

```python
new_df = merged_df.groupby([merged_df.Who_First, merged_df.depth, merged_df.Which_Heur, merged_df.move_number])["Prunings"].agg(["mean"])
new_df
## Get data for depths 4 and 5

## Now make a line graph of the means, do 2 graphs, each with 4 lines. Each graph should focus who_first, while 
## each line should focus which_heur at each depth
```
