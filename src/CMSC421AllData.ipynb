```python
import pandas as pd
pd.options.display.max_rows = 4000
# This gets Tyler's data
FILE_NAMES = ['../temp/exp3_d3_data.csv', '../temp/exp3_d4_data.csv', '../temp/exp3_d5_data.csv']
# This gets Alex's data
# FILE_NAMES = ['../Experiments/exp3_d3_data.csv', '../Experiments/exp3_d4_data.csv', '../Experiments/exp3_d5_data.csv']

df_list = []

for file_name in FILE_NAMES:
    df = pd.read_csv(file_name)
    df = df = df[df['depth'].notna()] # Remove empty rows 
    df_list.append(df)
    
merged_df = pd.concat(df_list, ignore_index=True)
merged_df.columns = [c.replace(' ', '_') for c in merged_df.columns]
merged_df
```

```python
newdf = merged_df.groupby([merged_df.depth, merged_df.move_number, merged_df.Going_First])["Minimax_Bot_time_to_Make_Move"].agg(["mean", "std"])
```

```python
newdf.to_csv("e3_tyler.csv")
```
